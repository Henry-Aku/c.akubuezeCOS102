```python
import numpy as np
import cmath

def solve_quadratic():
    print("\n--- Quadratic Equation: Ax^2 + Bx + C = 0 ---")
    a = float(input("Enter A: "))
    b = float(input("Enter B: "))
    c = float(input("Enter C: "))
    
    # Using the Quadratic Formula: (-b ± sqrt(b^2 - 4ac)) / 2a
    discriminant = (b**2) - (4*a*c)
    
    root1 = (-b + cmath.sqrt(discriminant)) / (2*a)
    root2 = (-b - cmath.sqrt(discriminant)) / (2*a)
    
    print(f"Roots: {root1} and {root2}")

def solve_polynomial(degree):
    names = {3: "Cubic", 4: "Quartic"}
    print(f"\n--- {names[degree]} Equation ---")
    
    # Collect coefficients from highest power to constant
    coeffs = []
    for i in range(degree, -1, -1):
        val = float(input(f"Enter coefficient for x^{i}: "))
        coeffs.append(val)
    
    # numpy.roots calculates roots for any polynomial degree
    roots = np.roots(coeffs)
    
    print(f"The roots of the {names[degree]} equation are:")
    for i, r in enumerate(roots, 1):
        print(f"Root {i}: {r}")

def main():
    while True:
        print("\n==============================")
        print("   POLYNOMIAL ROOT FINDER")
        print("==============================")
        print("1. Solve Quadratic (Degree 2)")
        print("2. Solve Cubic (Degree 3)")
        print("3. Solve Quartic (Degree 4)")
        print("4. Exit")
        
        choice = input("\nSelect an operation (1-4): ")
        
        if choice == '1':
            solve_quadratic()
        elif choice == '2':
            solve_polynomial(3)
        elif choice == '3':
            solve_polynomial(4)
        elif choice == '4':
            print("Mission accomplished. Exiting...")
            break
        else:
            print("Invalid selection. Please try again.")

if __name__ == "__main__":
    main()
```
