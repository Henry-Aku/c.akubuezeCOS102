```python
def calculate_atr():
    print("--- Izifin Technology ---")
    
    try:
        # Taking inputs from the user
        experience = int(input("Enter years of work experience: "))
        age = int(input("Enter age of the staff: "))

        # Conditional logic based on the requirements
        if experience > 25 and age >= 55:
            atr = 5600000
        elif experience > 20 and age >= 45:
            atr = 4480000
        elif experience > 10 and age >= 35:
            atr = 1500000
        else:
            # Covers staff < 10 years experience and < 35 years of age
            atr = 550000
            
        # Displaying the result formatted with commas for Naira
        print(f"\nThe Annual Tax Revenue (ATR) for this staff is: ₦{atr:,}")

    except ValueError:
        print("Invalid input. Please enter numerical values for age and experience.")

# Run the program
if __name__ == "__main__":
    calculate_atr()
```
