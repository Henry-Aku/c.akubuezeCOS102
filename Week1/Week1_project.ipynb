```python
def calculate_finances():
    print("Financial Calculator")
    
    # Common Inputs
    p = float(input("Enter Principal/Monthly Payment (P or PMT): "))
    r_percent = float(input("Enter Annual Interest Rate (as %): "))
    t = float(input("Enter Time (Years): "))
    n = float(input("Enter number of compounds per year (n): "))

    # Convert percentage to decimal
    r = r_percent / 100

    # a) Simple Interest (Corrected formula)
    # A = P * (1 + (R/100) * T)
    a_simple = p * (1 + r * t)

    # b) Compound Interest (Corrected formula)
    # A = P * (1 + R/n)^(n*t)
    a_compound = p * (1 + r/n)**(n * t)

    # c) Annuity Plan (Future Value of Ordinary Annuity)
    # A = PMT * [((1 + R/n)^(n*t) - 1) / (R/n)]
    if r > 0:
        a_annuity = p * (((1 + r/n)**(n * t) - 1) / (r/n))
    else:
        a_annuity = p * n * t # If interest is 0

    print("\n--- Results ---")
    print(f"a) Simple Interest Total:  ${a_simple:,.2f}")
    print(f"b) Compound Interest Total: ${a_compound:,.2f}")
    print(f"c) Annuity Plan Total:      ${a_annuity:,.2f}")

if __name__ == "__main__":
    calculate_finances()
```
